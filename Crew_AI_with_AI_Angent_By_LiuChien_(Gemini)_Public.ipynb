```python
# GitHub å°ˆæ¡ˆç¶²å€
github_url = "https://github.com/liuchiente/fonleeEnterprise.git"
project_name = github_url.split("/")[-1].replace(".git", "")
project_path = f"/content/{project_name}"

# Clone å°ˆæ¡ˆ
!git clone {github_url}
```

```python
import os, json
from rich.tree import Tree
from rich import print
import networkx as nx
import matplotlib.pyplot as plt

def build_tree(path, tree=None):
    if tree is None:
        tree = Tree(f"[bold blue]{os.path.basename(path)}[/]")
    for item in sorted(os.listdir(path)):
        full_path = os.path.join(path, item)
        if os.path.isdir(full_path):
            subtree = tree.add(f"[bold green]{item}/[/]")
            build_tree(full_path, subtree)
        else:
            tree.add(item)
    return tree

print("[bold underline]å°ˆæ¡ˆè³‡æ–™å¤¾çµæ§‹ï¼š[/]")
tree = build_tree(project_path)
print(tree)

# è§£æ package.json
print("[bold underline]ä¾è³´å¥—ä»¶çµæ§‹ï¼š[/]")
package_path = os.path.join(project_path, "package.json")
dependencies = {}
if os.path.exists(package_path):
    with open(package_path, "r") as f:
        package_data = json.load(f)
    dependencies = package_data.get("dependencies", {})
    for dep, version in dependencies.items():
        print(f"- {dep}: {version}")
```

```python
from crewai import Crew, Agent, Task
from langchain_community.llms import HuggingFaceHub

# Agent 1ï¼šçµæ§‹åˆ†æ
structure_agent = Agent(
    role="Structure Analyst",
    goal="åˆ†æå°ˆæ¡ˆè³‡æ–™å¤¾çµæ§‹ä¸¦æ‰¾å‡ºæ¨¡çµ„é‚è¼¯",
    backstory="ä½ æ˜¯è³‡æ·±æ¶æ§‹å¸«ï¼Œæ“…é•·å¾ç›®éŒ„çµæ§‹åˆ¤æ–·ç¨‹å¼è¨­è¨ˆæ¨¡å¼ã€‚",
    llm="gemini/gemini-2.5-flash"  # é€é LiteLLM è‡ªå‹•è™•ç†
)

# Agent 2ï¼šä¾è³´åˆ†æ
dependency_agent = Agent(
    role="Dependency Analyst",
    goal="è§£æ package.json ä¸¦æ‰¾å‡ºæ½›åœ¨ç›¸ä¾å•é¡Œ",
    backstory="ä½ æ˜¯ Node.js å°ˆå®¶ï¼Œç†Ÿæ‚‰å„ç¨®å¥—ä»¶ç‰ˆæœ¬èˆ‡ç›¸å®¹æ€§ã€‚",
    llm="gemini/gemini-2.5-flash"
)

# Agent 3ï¼šé‡æ§‹å»ºè­°
refactor_agent = Agent(
    role="Refactor Strategist",
    goal="æ ¹æ“šçµæ§‹èˆ‡ä¾è³´åˆ†æï¼Œæå‡ºé‡æ§‹å»ºè­°",
    backstory="ä½ æ˜¯å°ˆæ¡ˆé¡§å•ï¼Œå°ˆç²¾æ–¼æ¨¡çµ„åŒ–èˆ‡æŠ€è¡“å‚µæ¸…ç†ã€‚",
    llm="gemini/gemini-2.5-flash"
)
```

```python
# å°‡æƒæçµæœè½‰ç‚ºæ–‡å­—è¼¸å…¥
structure_text = tree.__str__()
dependency_text = "\n".join([f"{k}: {v}" for k, v in dependencies.items()])

task1 = Task(
    description=f"æ ¹æ“šä»¥ä¸‹è³‡æ–™å¤¾çµæ§‹åˆ†ææ¨¡çµ„é‚è¼¯ï¼š\n{structure_text}",
    expected_output="åˆ—å‡ºå°ˆæ¡ˆçš„ä¸»è¦æ¨¡çµ„èˆ‡å…¶åŠŸèƒ½å®šä½",
    agent=structure_agent
)

task2 = Task(
    description=f"æ ¹æ“šä»¥ä¸‹ dependencies åˆ†æç›¸ä¾å•é¡Œï¼š\n{dependency_text}",
    expected_output="æŒ‡å‡ºå¯èƒ½çš„ç‰ˆæœ¬è¡çªã€éæ™‚å¥—ä»¶æˆ–å®‰å…¨æ€§é¢¨éšª",
    agent=dependency_agent
)

task3 = Task(
    description="è«‹æ ¹æ“šå‰å…©é …åˆ†æçµæœï¼Œæå‡ºé‡æ§‹å»ºè­°èˆ‡æ¨¡çµ„å„ªåŒ–æ–¹å‘ã€‚",
    expected_output="æä¾›å…·é«”é‡æ§‹å»ºè­°èˆ‡æ¨¡çµ„æ‹†åˆ†ç­–ç•¥",
    agent=refactor_agent
)

# å»ºç«‹ä¸€å€‹ CrewAI çš„åœ˜éšŠï¼ˆCrewï¼‰ç‰©ä»¶
crew = Crew(
    agents=[structure_agent, dependency_agent, refactor_agent],  # æŒ‡å®šä¸‰ä½ Agentï¼Œå„è‡ªè² è²¬ä¸åŒåˆ†æä»»å‹™
    tasks=[task1, task2, task3],  # æŒ‡å®šä¸‰å€‹Taskï¼Œæœƒä¾ç…§é †åºåˆ†é…çµ¦ Agent åŸ·è¡Œ
    verbose=True  # å•Ÿç”¨è©³ç´°æ¨¡å¼ï¼ŒåŸ·è¡Œéç¨‹ä¸­æœƒå°å‡ºæ¯å€‹ Agent çš„æ€è€ƒèˆ‡å›è¦†
)

# å•Ÿå‹•æ•´å€‹åˆ†ææµç¨‹ï¼Œè®“ Agent é–‹å§‹åŸ·è¡Œä»»å‹™ä¸¦å”ä½œç”¢å‡ºçµæœ
result = crew.kickoff()
```
